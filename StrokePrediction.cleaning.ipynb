Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/healthcare-dataset-stroke-data.csv')
df
```

```python
df.info()
```

```python
df = df.drop(['id'], axis=1)
df['gender'] = df['gender'].replace({'Other' : 'Female'} )
df['smoking_status'] = df['smoking_status'].replace({'Unknown': 0})
```

```python
df
```

```python
df['ever_married'] = df['ever_married'].replace({'No': 0, 'Yes': 1})
df['gender'] = df['gender'].replace({'Female': 1, 'Male': 0})
df['Residence_type'] = df['Residence_type'].replace({'Rural': 0, 'Urban': 1})
df['work_type'] = df['work_type'].replace({'Private': 0, 'Self-employed': 1, 'Govt_job': 2, 'children': 3, 'Never_worked': 4})
df['smoking_status'] = df['smoking_status'].replace({'never smoked': 0, 'formerly smoked': 1, 'smokes': 2})
df['bmi'].interpolate(inplace=True)
df['age'] = df['age'].astype(int)
df
```

```python
df.info()
```

```python
df['gender']=df['gender'].astype(int)
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df.isnull().sum()
```

balancing the data

```python
from imblearn.over_sampling import SMOTE

X = df.drop('stroke', axis=1)
y = df['stroke']

smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Create a new DataFrame with the balanced data
balanced_df = pd.DataFrame(X_resampled)
balanced_df['stroke'] = y_resampled

# Save the balanced dataset to a new CSV file
balanced_df.to_csv('clean_data.csv', index=False)
```
